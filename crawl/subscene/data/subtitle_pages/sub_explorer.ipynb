```python
f = open('sub_pages.txt')
l = [tuple(x.strip().split()) for x in f]
```

```python
from collections import defaultdict
d = defaultdict(list)
for lan, _, title in l:
    d[lan].append(title)
```

```python
for x, y in d.iteritems():
    print x, len(y)
```

```python
rd = defaultdict(list)
for lan, _, title in l:
    rd[title].append(lan)
```

```python
languages = ['farsi_persian', 'english']

print sum(all(l in d[x] for l in languages) for x in d)
```
